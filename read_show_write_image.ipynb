# Show Image on Jupyter

```python
'''
This part of code works in Jupyter. We need matplotlib to show image. 

img = cv2.imread("lena.JPG", cv2.IMREAD_UNCHANGED)
the same as: 
img = cv2.imread("lena.JPG", 1)
1: cv2.IMREAD_COLOR : load a color image
0: load image in grayscale mode
-1: load image as such including alpha channel. 
'''
import cv2
from matplotlib import pyplot as plt

img = cv2.imread("lena.jpg", cv2.IMREAD_GRAYSCALE)

# Show the image with matplotlib
# as opencv loads in BGR format by default, we want to show it in RGB.
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) 
plt.imshow(img)
plt.show()
```

# Write image

```python
cv2.imwrite("lena_gray.png", img)
```
